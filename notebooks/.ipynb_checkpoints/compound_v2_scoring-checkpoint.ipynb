```python
# Import necessary libraries
import os
import json
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler

# Define the directory containing the JSON files
data_directory = 'data/'

# Define the filenames
filenames = [
    'compoundV2_transactions_ethereum_chunk_2.json',
    'compoundV2_transactions_ethereum_chunk_3.json',
    'compoundV2_transactions_ethereum_chunk_0.json'
]

# Function to load JSON files and flatten nested fields
def load_json_files(directory, filenames):
    data_frames = []
    for filename in filenames:
        file_path = os.path.join(directory, filename)
        with open(file_path, 'r') as file:
            data = json.load(file)
        df = pd.json_normalize(
            data['deposits'],
            sep='.',
            record_prefix='',
            meta=[
                ['account', 'id'],
                ['asset', 'id'],
                ['asset', 'symbol'],
                'amount',
                'amountUSD',
                'hash',
                'id',
                'timestamp'
            ]
        )
        data_frames.append(df)
    return pd.concat(data_frames, ignore_index=True)

# Load the data
data = load_json_files(data_directory, filenames)

# Rename nested columns if needed
data.rename(columns={
    'account.id': 'account_id',
    'asset.id': 'asset_id',
    'asset.symbol': 'asset_symbol'
}, inplace=True)

# Display basic information about the dataset
print("Initial Data Info:")
print(data.info())
print(data.describe())

# Convert numerical columns to appropriate types
data['amount'] = pd.to_numeric(data['amount'], errors='coerce')
data['amountUSD'] = pd.to_numeric(data['amountUSD'], errors='coerce')
data['timestamp'] = pd.to_datetime(data['timestamp'], unit='s', errors='coerce')

# Handle missing values
data.dropna(subset=['amount', 'amountUSD', 'timestamp'], inplace=True)

# Set timestamp as index
data.set_index('timestamp', inplace=True)

# Print data after cleaning
print("Data after cleaning:")
print(data.info())
print(data.describe())

# Feature Engineering
# Check required column
if 'account_id' not in data.columns:
    raise ValueError("Column 'account_id' is missing in the data.")
if 'amount' not in data.columns:
    raise ValueError("Column 'amount' is missing in the data.")

# Perform aggregation
wallet_features = data.groupby('account_id').agg(
    total_deposits=('amount', 'sum'),
    transaction_frequency=('amount', 'count'),
    average_transaction=('amount', 'mean'),
    max_transaction=('amount', 'max'),
    min_transaction=('amount', 'min')
).reset_index()

# Ensure the DataFrame is not empty
if wallet_features.empty:
    raise ValueError("Feature engineering resulted in an empty DataFrame.")

# Apply logarithmic transformations
wallet_features['log_total_deposits'] = wallet_features['total_deposits'].apply(lambda x: np.log(x + 1))
wallet_features['log_average_transaction'] = wallet_features['average_transaction'].apply(lambda x: np.log(x + 1))

# Model Development
scaler = StandardScaler()
scaled_features = scaler.fit_transform(wallet_features.drop(columns=['account_id']))

kmeans = KMeans(n_clusters=5, random_state=42, n_init=10)
wallet_features['cluster'] = kmeans.fit_predict(scaled_features)

# Assign scores based on clusters
wallet_features['score'] = wallet_features['cluster'].map({0: 100, 1: 80, 2: 60, 3: 40, 4: 20})

# Scoring and Output
top_1000_wallets = wallet_features.nlargest(1000, 'score')
os.makedirs('output', exist_ok=True)
top_1000_wallets[['account_id', 'score']].to_csv('output/wallet_scores.csv', index=False)

# Wallet Analysis
high_scoring_wallets = top_1000_wallets.head(5)
low_scoring_wallets = top_1000_wallets.tail(5)

print("High Scoring Wallets Analysis:")
print(high_scoring_wallets)

print("\nLow Scoring Wallets Analysis:")
print(low_scoring_wallets)
```

```python
# Visualization of clusters
plt.figure(figsize=(10, 6))
sns.scatterplot(
    x='log_total_deposits',
    y='log_average_transaction',
    hue='cluster',
    data=wallet_features,
    palette='viridis',
    style='cluster',
    markers=True,
    s=100
)
plt.title('Cluster Analysis')
plt.xlabel('Log Total Deposits')
plt.ylabel('Log Average Transaction')
plt.legend(title='Cluster')
plt.show()
```

```python
# Calculate summary statistics for each cluster
cluster_stats = wallet_features.groupby('cluster').describe()
print(cluster_stats)
```

```python
# Function to get top wallets in a cluster
def get_top_wallets(cluster, num_top_wallets=5):
    top_wallets = wallet_features[wallet_features['cluster'] == cluster].nlargest(num_top_wallets, 'score')
    return top_wallets[['account_id', 'score', 'total_deposits', 'transaction_frequency']]

# Get top wallets in each cluster
for cluster_id in wallet_features['cluster'].unique():
    print(f"Top wallets in cluster {cluster_id}:")
    print(get_top_wallets(cluster_id))
    print("\n")
```

```python
import scipy.stats as stats

# Compare the means of 'total_deposits' across clusters
anova_results = stats.f_oneway(
    *(wallet_features[wallet_features['cluster'] == i]['total_deposits'] for i in wallet_features['cluster'].unique())
)
print(f"ANOVA results for total_deposits: F={anova_results.statistic}, p-value={anova_results.pvalue}")
```

```python
import os
import json
import pandas as pd

# Function to extract nested records
def extract_actions_from_nested_json(json_data, source_file):
    extracted_rows = []

    for action_type in ['deposits', 'withdraws', 'borrows', 'repays', 'liquidates']:
        if action_type in json_data:
            records = json_data[action_type]
            for record in records:
                record_flat = record.copy()
                record_flat['action'] = action_type
                record_flat['source_file'] = source_file
                extracted_rows.append(record_flat)

    return extracted_rows

# Function to load and process files
def load_and_extract_all_actions(directory, filenames):
    all_data = []

    for file in filenames:
        file_path = os.path.join(directory, file)
        try:
            if os.path.exists(file_path):
                with open(file_path, 'r') as f:
                    json_data = json.load(f)
                    extracted = extract_actions_from_nested_json(json_data, file)
                    if extracted:
                        all_data.extend(extracted)
                        print(f" Extracted {len(extracted)} records from {file}")
                    else:
                        print(f" No records found in {file}")
            else:
                print(f" File not found: {file_path}")
        except Exception as e:
            print(f" Error loading {file}: {e}")

    # Convert to DataFrame
    if all_data:
        df = pd.DataFrame(all_data)
    else:
        df = pd.DataFrame()

    return df

# Feature Engineering Function
def feature_engineering(df):
    if 'account_id' not in df.columns or 'action' not in df.columns:
        print(" Required columns missing. Cannot perform feature engineering.")
        return None

    group = df.groupby(['account_id', 'action'])
    features = group['amountUSD'].agg(['count', 'sum', 'mean', 'max', 'min']).unstack().fillna(0)
    features.columns = ['_'.join(col).lower() for col in features.columns]
    return features

# Define directory and filenames
data_directory = 'data'  # Make sure the folder is named 'data' and is in the same directory as the script
filenames = [
    'compoundV2_transactions_ethereum_chunk_0.json',
    'compoundV2_transactions_ethereum_chunk_2.json',
    'compoundV2_transactions_ethereum_chunk_3.json'
]

# Load data
df = load_and_extract_all_actions(data_directory, filenames)

if not df.empty:
    print(f" Loaded and flattened total {len(df)} rows.")
    if 'account_id' not in df.columns:
        print(" 'account_id' column still missing. Printing available columns for reference:")
        print(df.columns.tolist())
    else:
        features = feature_engineering(df)
        if features is not None:
            print(" Feature engineering result:")
            print(features.head())
else:
    print(" No data extracted. Please inspect JSON structure or file contents.")

```

```python
def feature_engineering(df):
    # Ensure correct data types
    if isinstance(df['account_id'].iloc[0], dict):
        df['account_id'] = df['account_id'].apply(lambda x: x.get('id') if isinstance(x, dict) else x)
    if isinstance(df['action'].iloc[0], dict):
        df['action'] = df['action'].apply(lambda x: x.get('type') if isinstance(x, dict) else x)

    # Ensure both are strings for grouping
    df['account_id'] = df['account_id'].astype(str)
    df['action'] = df['action'].astype(str)

    # Now proceed to groupby
    group = df.groupby(['account_id', 'action'])
    features = group['amountUSD'].agg(['count', 'sum', 'mean', 'max', 'min']).unstack().fillna(0)
    features.columns = ['_'.join(col).lower() for col in features.columns]
    return features
```

```python
import pandas as pd
import json
import glob

def load_and_flatten_files(path_pattern):
    all_dfs = []
    for file in glob.glob(path_pattern):
        with open(file, 'r') as f:
            data = json.load(f)
            df = pd.json_normalize(data)
            print(f" Extracted {len(df)} records from {file}")
            all_dfs.append(df)
    combined_df = pd.concat(all_dfs, ignore_index=True)
    print(f" Loaded and flattened total {len(combined_df)} rows.")
    return combined_df

# Step 1: Load raw nested data
combined_df = load_and_flatten_files("compoundV2_transactions_ethereum_chunk_*.json")

# Step 2: Flatten each transaction type and extract account_id
def extract_event_df(event_name):
    records = []
    for i, row in combined_df.iterrows():
        if isinstance(row[event_name], list):
            for entry in row[event_name]:
                if 'account' in entry and 'id' in entry['account']:
                    records.append({
                        'account_id': entry['account']['id'],
                        'event_type': event_name,
                        **entry
                    })
    return pd.DataFrame(records)

# Step 3: Merge all event DataFrames
events = ['deposits', 'withdraws', 'borrows', 'repays']
final_df = pd.concat([extract_event_df(event) for event in events], ignore_index=True)

# Step 4: Drop duplicate 'account' dict from each row
if 'account' in final_df.columns:
    final_df.drop(columns=['account'], inplace=True)

# Step 5: Show results
print(" Final flattened DataFrame with account_id:")
print(final_df.head())

# Optional: Save to CSV
# final_df.to_csv("flattened_compound_events.csv", index=False)
```

```python
# Count how many events each account participated in
event_counts = final_df.groupby('account_id')['event_type'].count().reset_index(name='total_events')
print(event_counts.head())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Count of each event type
plt.figure(figsize=(6, 4))
sns.countplot(data=final_df, x='event_type', order=final_df['event_type'].value_counts().index)
plt.title("Distribution of Event Types")
plt.xlabel("Event Type")
plt.ylabel("Count")
plt.tight_layout()
plt.show()
```

```python
top_accounts = final_df[final_df['event_type'] == 'deposits'].copy()
top_accounts['amountUSD'] = pd.to_numeric(top_accounts['amountUSD'], errors='coerce')

top10 = top_accounts.groupby('account_id')['amountUSD'].sum().nlargest(10).reset_index()

plt.figure(figsize=(10, 6))
sns.barplot(data=top10, x='amountUSD', y='account_id', palette='viridis')
plt.title("Top 10 Accounts by Total USD Deposited")
plt.xlabel("Total Deposited (USD)")
plt.ylabel("Account ID")
plt.tight_layout()
plt.show()
```

```python
# Convert UNIX timestamp to datetime
final_df['datetime'] = pd.to_datetime(final_df['timestamp'], unit='s')

# Filter only deposits and aggregate over time
deposits_ts = final_df[final_df['event_type'] == 'deposits'].copy()
deposits_ts['amountUSD'] = pd.to_numeric(deposits_ts['amountUSD'], errors='coerce')

# Resample to daily total
daily_deposits = deposits_ts.set_index('datetime').resample('D')['amountUSD'].sum().reset_index()

plt.figure(figsize=(12, 5))
sns.lineplot(data=daily_deposits, x='datetime', y='amountUSD')
plt.title("Daily Total Deposits (USD)")
plt.xlabel("Date")
plt.ylabel("Total Deposits in USD")
plt.tight_layout()
plt.show()
```
